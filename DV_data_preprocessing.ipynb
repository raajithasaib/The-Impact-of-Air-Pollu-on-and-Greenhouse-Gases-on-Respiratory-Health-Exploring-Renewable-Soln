```python
import pandas as pd
import os

# List of file paths for each CSV file
file_paths = [
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Carbon dioxide (CO2) Emissions without Land Use, Land-Use Change and Forestry (LULUCF), in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Greenhouse Gas (GHGs) Emissions without Land Use, Land-Use Change and Forestry (LULUCF), in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Greenhouse Gas (GHGs) Emissions, including Indirect CO2, without LULUCF, in kilotonne CO2 equivalent .csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Hydrofluorocarbons (HFCs) Emissions, in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Methane (CH4) Emissions without Land Use, Land-Use Change and Forestry (LULUCF), in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Nitrogen trifluoride (NF3) Emissions, in kilotonne CO2 equivalent .csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Nitrous oxide (N2O) Emissions without Land Use, Land-Use Change and Forestry (LULUCF), in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Perfluorocarbons (PFCs) Emissions, in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Sulphur hexafluoride (SF6) Emissions, in kilotonne CO2 equivalent.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/UNdata_Unspecified mix of Hydrofluorocarbons (HFCs) and Perfluorocarbons (PFCs) emissions, in kilotonne CO2 equivalent .csv"
]

# Initialize a variable to hold the merged DataFrame
combined_df = None

# Loop through each file path
for file in file_paths:
    # Extract the file name without extension
    emission_type = os.path.splitext(os.path.basename(file))[0]
    
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Rename the 'Value' column to the emission type (formatted)
    df = df.rename(columns={'Value': emission_type.replace("_", " ").title()})
    
    # Merge with the combined_df on 'Country or Area' and 'Year'
    if combined_df is None:
        # Initialize combined_df with the first file's data
        combined_df = df
    else:
        # Merge subsequent files on 'Country or Area' and 'Year'
        combined_df = pd.merge(combined_df, df, on=['Country or Area', 'Year'], how='outer')

# Sort the combined DataFrame by 'Year' (and 'Country or Area' if needed)
combined_df = combined_df.sort_values(by=['Year', 'Country or Area']).reset_index(drop=True)

# Display the final combined and sorted DataFrame
print(combined_df.head())
```

```python
# Save the final combined and sorted DataFrame to an Excel file
output_file_path = "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/Combined_UN_Data.xlsx"
combined_df.to_excel(output_file_path, index=False)

print(f"Combined data has been saved to {output_file_path}")
```

```python
import pandas as pd

# Define file paths
file_paths = [
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/per-capita-ghg-emissions.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/per-capita-methane-emissions.csv",
    "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/per-capita-nitrous-oxide.csv"
]

# Initialize an empty DataFrame to hold the combined data
combined_df = None

# Loop through each file and merge them
for file in file_paths:
    # Extract the file name without extension to create unique column names
    emission_type = file.split('/')[-1].split('.')[0]
    
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Rename the relevant 'Value' column to include the emission type for differentiation
    # Assuming that each file has a single relevant column with the emissions value
    df.columns = ['Entity', 'Code', 'Year', emission_type]
    
    # Merge with the combined DataFrame on 'Entity' and 'Year'
    if combined_df is None:
        # Initialize combined_df with the first file's data
        combined_df = df
    else:
        # Merge subsequent files on 'Entity' and 'Year'
        combined_df = pd.merge(combined_df, df, on=['Entity', 'Year'], how='outer')

# Display the final combined DataFrame
print(combined_df.head())

# Save the combined DataFrame to an Excel file if needed
output_file_path = "/Users/raajithasai/Documents/Semester 3/Data Visualisation/Final Project/Data Collection/Combined_Emissions_Data_per_capita.xlsx"
combined_df.to_excel(output_file_path, index=False)
print(f"Combined data has been saved to {output_file_path}")
```
